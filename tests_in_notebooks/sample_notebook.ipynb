## Setup & Imports

First, let's import the necessary modules from dataengineer_toolbox.

## Test 1: Basic VTableModel Creation

Create a simple VTableModel instance and display its properties.

```python
# Example: Create a VTableModel
model = VTableModel(
    name="customer_vtable",
    description="Customer virtual table model"
)

print(f"Model Name: {model.name}")
print(f"Model Description: {model.description}")
```

## Test 2: Entity Working

Create and work with Entity objects.

```python
# Create an Entity
entity = Entity(
    name="customer",
    table_name="customers"
)

print(f"Entity Name: {entity.name}")
print(f"Entity Table Name: {entity.table_name}")
```

## Test 4: Serialization Test

Test model serialization capabilities.

```python
# Test serialization
try:
    model_dict = model.dict() if hasattr(model, 'dict') else vars(model)
    print("Model serialized successfully:")
    print(model_dict)
except Exception as e:
    print(f"Error during serialization: {e}")
```
